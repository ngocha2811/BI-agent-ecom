```python
import pandas as pd
```

```python
data = pd.read_csv('path')
```

```python
# Map values in traitement column, Vendeur -> Seller, Amazon -> Amazon
data['traitement'] = data['traitement'].map({'Vendeur': 'Seller', 'Amazon': 'Amazon'})

# change column name to type
data.rename(columns={'traitement': 'type'}, inplace=True)
```

```python
# Change column names from French to English
data.rename(columns={'date/heure': 'date', 'ventes_de_produits': 'sale_price','total_frais' : 'fee'}, inplace=True)
```

```python
# Drop column numéro_de_versement
data.drop(columns=['numéro_de_versement'], inplace=True)
```

```python
data.head()
```

```python
# Map values in type column: Commande -> Order, Retour -> Return
data['type'] = data['type'].map({'Commande': 'Order', 'Retour': 'Return'})
data.head()
```

```python
# Save data to csv
data.to_csv('path', index=False)
```
